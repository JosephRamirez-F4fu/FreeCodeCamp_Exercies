```python
%%writefile declaration.hpp
#include <iostream>
#include <ctime>
#include <cstdlib>
using std::cout;
using std::cin;
using std::endl;

template<typename T>
T suma(T a,T b);

template<typename T>
T encontrar_mayor_arreglo(T* arr, int N);

template<typename T>
T encontrar_menor_arreglo(T* arr, int N);

template<typename T>
T sumar_todos_arreglo(T* arr, int N);

template<typename T>
void moda_arreglo(T* arr, int N);
```

```python
%%writefile implementation.hpp
#include "declaration.hpp"
template<typename T>
T suma(T a,T b) {
	return a + b;
}

template<typename T>
T encontrar_mayor_arreglo(T* arr, int N) {
	/*Esta funcion tipo tenplate retorna el maximo valor del arreglo del tipo T 
	Requiriendo el arreglo,la la longitud del arreglo*/
	T aux = arr[0];
	for (int i = 1; i < N; i++)
	{
		if (aux <= arr[i])
			aux = arr[i];
	}
	
	return aux;
}

template<typename T>
T encontrar_menor_arreglo(T* arr, int N) {
	/*Esta funcion tipo tenplate retorna el menor valor del arreglo del tipo T
	Requiriendo el arreglo,la la longitud del arreglo*/
	T aux = arr[0];
	for (int i = 1; i < N; i++)
	{
		if (aux >= arr[i])
			aux = arr[i];
	}

	return aux;
}

template<typename T>
T sumar_todos_arreglo(T* arr, int N) {
	/*Esta funcion tipo tenplate retornar la suma de elementos del arreglo del arreglo del tipo T
	Requiriendo el arreglo,la la longitud del arreglo*/
	T aux = arr[0];
	for (int i = 1; i < N; i++)
	{
		aux += arr[i];
	}

	return aux;
}


template<typename T>
void moda_arreglo(T* arr, int N) {
	/*Esta funcion tipo tenplate retornar la moda de elementos del arreglo del arreglo del tipo T
	Requiriendo el arreglo,la la longitud del arreglo*/
	int rep=0, moda=arr[0],moda_rep=0;

	for (int i = 0; i < N ; i++)
	{	
		rep = 1;
		for (int j = i+1; j < N; j++)
		{
			if (arr[i] == arr[j]) {
				rep++;
			}
		}

		if (rep > moda_rep) {
			moda_rep = rep;
			moda = arr[i];
		}
		
	}
	if (moda_rep <= 1) {
		cout << "no hay moda";
	}
	else {
		cout << "La moda del arreglo es " << moda << " y se repite " << moda_rep <<" veces";
	}
}
```

```python
%%writefile execute_implementation.cpp
#include "implementation.hpp"
int main() {

	cout << suma<float>(3.2, 48)<<"\n";
	cout << suma<std::string>("hola ", "mundo");
	cout << endl;
	int arr[] = { 0,100,4,8,1,7,2,8,3,15,20,21,47,0,1,2,4,1,4,1,1,4,4 };
	cout<<encontrar_mayor_arreglo<int>(arr, sizeof arr / sizeof arr[0]);
	cout << endl;
	cout << encontrar_menor_arreglo<int>(arr, sizeof arr / sizeof arr[0]);
	cout << endl;
	cout << sumar_todos_arreglo<int>(arr, sizeof arr / sizeof arr[0]);
	cout << endl;
	moda_arreglo<int>(arr, sizeof arr / sizeof arr[0]);
	return 0;
}
```

##Class templates 

```python
%%writefile arreglo_class_declaration.hpp
#include <iostream>
#include <ctime>
#include <cstdlib>
using std::cout;
using std::cin;
using std::endl;

template<typename Generic>

class Arreglo(){
    Generic * a;
    size_t n;
    public:
    Arreglo(Generic *_a,site_t _n){a*=_a;n=_n;}
    Arreglo(){a=new Generic[10];n=0;}

    void insertar(Generic g){
        if(n<10)
          a(n++)=g;
    }
    void listar(){
        for(size_t i=0;i<n;++i){
            cout<<i<<"->"a[i]<<"\n";
        }
    }

};
```

```python
%%writefile arreglo_class_execute.cpp

int main(){

  Arreglo<int>*arregloenteros=new Arreglo<int>();
  Arreglo<float>*arregloflotantes=new Arreglo<float>();

  for(int i=0;i<10,i++)
  {
      arregloenteros.insertar(i*12)
  }
  arregloenteros.listar();
  return 0;
}
```
